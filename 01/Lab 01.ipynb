# **Task 3:**

#### Graphical visualization:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns

dataset[numeric_cols].hist(figsize=(15, 12), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=dataset[["Age", "Obesity", "Alcohol use"]])
plt.title("Boxplots of Selected Features")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=dataset, palette="Set2")
plt.title("Distribution of Gender")
plt.show()
```

#### Correlation heatmap:

```python
sns.heatmap(dataset[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

#### Disribution of age and frequent cold:

```python
plt.figure(figsize=(8,4))
sns.histplot(dataset["Age"], bins=20, color="teal")
plt.title("Distribution of Age")
plt.show()

# Distribution of Frequent Cold
plt.figure(figsize=(8,4))
sns.countplot(x="Frequent Cold", data=dataset, palette="pastel")
plt.title("Distribution of Frequent Cold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=dataset["Obesity"], color="coral")
plt.title("Outliers in Obesity Feature")
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(dataset["Obesity"], bins=15, color="orange")
plt.title("Distribution of Obesity Feature")
plt.show()
```
